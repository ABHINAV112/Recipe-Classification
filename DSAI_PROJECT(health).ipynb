```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
demo_df = pd.read_csv('demographics.csv')
demo_df.head()
```

```python
demo_df = pd.read_csv('demographics.csv', usecols=["SEQN","RIAGENDR","RIDAGEYR","RIDRETH1","RIDRETH3"])
demo_df.head()
```

```python
demo_df = demo_df.rename(columns = {"SEQN": "RESPONDENT", "RIAGENDR":"GENDER", "RIDAGEYR": "AGE","RIDRETH1":"RACE HISPANIC",
                                    "RIDRETH3":"RACE NON HISPANIC"})
demo_df.head()
```

```python
demo_clean = demo_df.copy()
demo_clean.columns = demo_clean.columns.str.replace(".","")
demo_clean.columns = demo_clean.columns.str.replace(" ","_")


demo_clean.info()
```

```python
demo_clean.isnull().sum()
```

```python
print(demo_clean["GENDER"].dropna().value_counts())
```

```python
labs_df = pd.read_csv('labs dataset.csv')
labs_df.head()
```

```python
labs_df= pd.read_csv('labs dataset.csv', usecols=["LBDHDD","LBDHDDSI","LBDTCSI","LBXTC","LBDSCASI"])
labs_df.head()
```

```python
labs_df = labs_df.rename(columns = { "LBDHDD":"HDL CHOL(mg/dL)", "LBDHDDSI": "HDL CHOL(mmol/L)","LBDTCSI":"TOTAL CHOL(mmol/L)",
                                    "LBXTC":"TOTAL CHOL","LBDSCASI":"TOT CALUCIUM(mmol/L)"})
labs_df.head()
```

```python
labs_df.isnull().sum()
```

```python
labs_df["TOT CALUCIUM(mmol/L)"].fillna(value = "NoType", inplace = True)
labs_df["HDL CHOL(mg/dL)"].fillna(value = "NoType", inplace = True)
labs_df["HDL CHOL(mmol/L)"].fillna(value = "NoType", inplace = True)
labs_df["TOTAL CHOL"].fillna(value = "NoType", inplace = True)
labs_df["TOTAL CHOL(mmol/L)"].fillna(value = "NoType", inplace = True)
labs_df.info()
```

```python
diet_df = pd.read_csv('diet.csv')
diet_df.head()
```

```python
diet_df = pd.read_csv('diet.csv',usecols=["DR1TNUMF","DR1TALCO","DR1TCALC","DR1TCHOL","DR1TCARB","DR1TCAFF",
                                          "DR1TKCAL","DR1TPFAT","DR1TPROT","DR1TSODI","DR1TSUGR",
                                          "DR1TTFAT","DRQSDIET"])
diet_df.head()
```

```python
diet_df = diet_df.rename(columns = { "DRQSDIET":"ANY DIET?","DR1TNUMF":"FOOD/BEV COUNT","DR1TTFAT":"TOTAL FAT(gm)",
                                    "DR1TKCAL":"ENERGY(kcal)","DR1TPROT":"PROTEIN(gm)","DR1TCARB":"CARBS(gm)","DR1TSUGR":"TOTAL SUGARS(gm)",
                                   "DR1TPFAT":"POLYUNSAT. FATS(gm)","DR1TCHOL":"CHOLESTROL(mg)","DR1TCALC":"CALCIUM(mg)","DR1TSODI":"SODIUM(mg)","DR1TCAFF":"CAFFEINE(mg)","DR1TALCO":"ALCOHOL(gm)"})


diet_df.head()
```

```python
diet_df.isnull().sum()
```

```python
diet_df["ANY DIET?"].fillna(value = "NoType", inplace = True)
diet_df["FOOD/BEV COUNT"].fillna(value = "NoType", inplace = True)
diet_df["ENERGY(kcal)"].fillna(value = "NoType", inplace = True)
diet_df["PROTEIN(gm)"].fillna(value = "NoType", inplace = True)
diet_df["CARBS(gm)"].fillna(value = "NoType", inplace = True)
diet_df["TOTAL SUGARS(gm)"].fillna(value = "NoType", inplace = True)
diet_df["TOTAL FAT(gm)"].fillna(value = "NoType", inplace = True)
diet_df["POLYUNSAT. FATS(gm)"].fillna(value = "NoType", inplace = True)
diet_df["CHOLESTROL(mg)"].fillna(value = "NoType", inplace = True)
diet_df["CALCIUM(mg)"].fillna(value = "NoType", inplace = True)
diet_df["SODIUM(mg)"].fillna(value = "NoType", inplace = True)
diet_df["CAFFEINE(mg)"].fillna(value = "NoType", inplace = True)
diet_df["ALCOHOL(gm)"].fillna(value = "NoType", inplace = True)

diet_df.info()
```

```python
healthDF = pd.concat([demo_clean, labs_df, diet_df], axis = 1, join_axes = [diet_df.index])
healthDF.head()
```

```python
healthDF = healthDF.set_index('RESPONDENT')
healthDF.head()
```

```python
healthDF.columns = healthDF.columns.str.replace(".","")
healthDF.columns = healthDF.columns.str.replace(" ","_")


healthDF.info()
```
